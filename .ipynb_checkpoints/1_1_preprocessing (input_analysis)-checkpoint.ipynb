```python
# Reading the metadata files for all satellite images in the data_dir folder

ds_read = xr.open_mfdataset(os.path.join(params['data_dir'],'*'), chunks='auto')
ds_read
```

```python
# Function for extracting the overall interesed region (e.g. whole atlantic ocean)

def ds_crop(ds_read, params):
    ds = ds_read.where(
        (ds_read.lat > params['boundaries']['lat_min'])&
        (ds_read.lon > params['boundaries']['lon_min'])&
        (ds_read.lat < params['boundaries']['lat_max'])&
        (ds_read.lon < params['boundaries']['lon_max']),
        drop=True
    ).sel(time=slice(params['start_date'],params['end_date']))
    
    flags = ds[params['flags']]
    ds_unc_rel= ds[params['UNC']]
    ds = ds[params['PFT']]
    
    return ds, ds_unc_rel, flags
```

# Uncertainty

Converting the relative uncertainty in linear scale to logscale standard deviation (SD) uncertainty

```python
# An empirical function that convert the relative uncertainty in linear scale to standard deviation uncertainty in logscale (based on )

def Unc_LinRel_to_LogSD(ds_unc_rel):
    ds_unc_sd = np.log10((ds_unc_rel/100)+1)
    return ds_unc_sd
```

```python
# renaming uncertainty to match with the ds
ds_unc_rel = ds_unc_rel.rename(params['UNC_dict'])

# converting the relative uncertainty to SD
ds_unc_sd = Unc_LinRel_to_LogSD(ds_unc_rel)

# Temporal mean of the uncertainty
ds_unc_rel_mean = ds_unc_rel.mean(dim='time')
ds_unc_sd_mean = ds_unc_sd.mean(dim='time')
```

# Missing rates

## Temporal

```python
def temporal_missing_rate(ds, rolling_windows_size=30):
    
    # Count the number of non-missing pixels per day for CHL and DIATO (as representative for all PFTs)
    temporal_count = ds[['CHL','DIATO']].count(dim=['lat','lon'])
    
    # Count the total number of sea pixels in the data
    sea_count = flags.isel(time=0).where(flags.isel(time=0)==0).count()
    
    # Daily percentage of missing values
    temporal_perc = 100 - (temporal_count / sea_count * 100)
    
    # Moving average daily percentage of missing values
    temporal_perc_mov = temporal_perc.rolling(time=rolling_windows_size, center=True).mean()

    return temporal_perc, temporal_perc_mov
```

```python
temporal_perc, temporal_perc_mov = temporal_missing_rate(ds,rolling_windows_size=30)
# temporal_perc.to_netcdf(os.path.join(params['work_dir'],'temporal_missing_perc.nc'))
# temporal_perc_mov.to_netcdf(os.path.join(params['work_dir'],'temporal_missing_perc_mov.nc'))
```

## Spatial

```python
def spatial_missing_rate(ds):
    
    # Count the number of non-missing pixels per location for CHL and DIATO (as representative for all PFTs)
    spatial_count = ds[['CHL','DIATO']].count(dim=['time'])
    
    # Mask for removing the land pixels
    msk = flags.isel(time=0, drop=True)
    
    # Compute the percentage of spatial missing rate
    spatial_perc = (100 - (spatial_count / ds.time.size * 100)).where(msk==0)
    
    return spatial_perc
```

```python
spatial_perc = spatial_missing_rate(ds)
# spatial_perc.to_netcdf(os.path.join(params['work_dir'],'spatial_missing_perc.nc'))
```
