# Functions

```python
def regression_analysis(p,x,y,x_log10, y_log10,f):
    ''' 
    This function recieve the satellite and in-situ measurements 
    and perform a regression analysis
    
    Parameters:
    p (str): Name of the Chla concentration data (PFT or TChla names)
    x (list): in-situ measurements [Linear scale]
    y (list): satellite observation [Linear scale]
    x_log10 (list): in-situ measurements [Logarithmic scale]
    y_log10 (list): satellite observation [Logarithmic scale]
    f (list): list of valid points that are finite in in-situ and satellite data
    
    Returns:
    m (float): slope of the regression
    b (float): intercept of the regression
    r2 (float): coefficient of determination of the regression
    fit_x (list): 100 x values for plotting regression line
    fit_y (list): 100 y values for plotting regression line
    y_pred (list): predicted y values based on x
    '''
    
    # General linear regression for TChla
    if p == 'CHL':
        result = linregress(x_log10[f], y_log10[f])
        m = result.slope
        b = result.intercept
        r2 = result.rvalue**2 
    
    else:
    # Type II linear regression for PFTs
        result = regress2(x_log10[f], y_log10[f], _method_type_2="major axis")
        m = result['slope']
        b = result['intercept']
        r2 = (result['r'])**2
    
    ## Create the ploynomial for regression line
    poly_func = np.poly1d((m, b))
    y_pred = poly_func(x_log10[f])    
    fit_x = np.linspace(0.001, 10, 1000)
    fit_y = 10**(poly_func(np.log10(fit_x)))
                              
    return m, b, r2, fit_x, fit_y, y_pred
```

```python
def add_95prediction_line(ax, y_pred, y_log10, fit_x, fit_y, color):
    ''' 
    This function adds the 95% prediction band to the plot.
    
    Parameters:
    ax (plt.ax): axis to add the prediction band
    y_pred (list): predicted y values based on x
    y_log10 (list): satellite observation [Logarithmic scale]
    fit_x (list): 100 x values for plotting regression line
    fit_y (list): 100 y values for plotting regression line
    color (str): color of the prediction band

    Returns:
    fit_line (matplotlib.lines.Line2D): Fitted linear regression line
    predict95up (matplotlib.lines.Line2D): upper 95% prediction band
    predict95dw (matplotlib.lines.Line2D): lower 95% prediction band
    '''
    
    std = np.std(y_pred - y_log10)
    z_score = 1.96
    y_lower = 10**(np.log10(fit_y) - z_score * std)
    y_upper = 10**(np.log10(fit_y) + z_score * std)
    fit_line, = ax.plot(fit_x, fit_y,  color=color,label='Fitted line',linewidth=1, alpha=1)
    predict95up, = ax.plot(fit_x, y_lower, color=color, linestyle=(5, (10, 3)), linewidth=1, alpha=0.5)
    predict95dw, = ax.plot(fit_x, y_upper, color=color, linestyle=(5, (10, 3)), linewidth=1, alpha=0.5, label=r'95\% prediction band')
    
    return fit_line, predict95up, predict95dw
```

```python
def matchups_number(ax, f, height, color):
    ''' 
    This function adds the number of matchups to the plot.
    
    Parameters:
    ax (plt.ax): axis to add the prediction band
    f (list): list of valid points that are finite in in-situ and satellite data
    height (int or float): The height to add the stats
    color (str): color for writing the matchup number stats

    Returns:
    None
    adds the text directly to the plot
    '''

    ax.text(9, height, r'$\mathbf{'+f'n={np.sum(f):.0f}'+'}$', ha='right', color=color, fontsize=14)
```

```python
def create_empty_stats_ds(params):
    ''' 
    This function creates an empy dataset to be filled with matchup statistics

    Parameters:
    params (dict): Dictionary of parameters

    Returns:
    dataset (xr.Dataset): empty dataset to store the regression analysis stasts.
    '''
    
    data = [
        "original", 
        "DINEOF[all]", 
        "DINEOF[transferred]", 
        "DINEOF[filled]", 
        "DINCAE[all]", 
        "DINCAE[transferred]", 
        "DINCAE[filled]"
    ]

    stats = [
        "N. matchups", 
        "Slope", 
        "Intercept", 
        "R2", 
        "MedPD", 
        "RMSD", 
        "NRMSD"
    ]


    dataset = xr.Dataset(
        {var: (["data", "stats"], np.full((len(data), len(stats)), np.nan)) for var in params['PFT']},
        coords={"data": data, "stats": stats},
    )
    return dataset
```

```python
## Function to extract the number of finite matchups in 
## in-situ and satellite measurements

lambda_fn = lambda y_log10, x_log10: [
    a and b for a, b in zip(np.isfinite(y_log10), np.isfinite(x_log10))
]
```

```python
def RMSD(x, y):
    ''' 
    Function to compute the root-mean-squared-deviation (RMSD)
    
    Parameters:
    x (list): in-situ measurements [Linear scale]
    y (list): satellite observation [Linear scale]
    
    Returns:
    RMSD (float): RMSD value between in-situ measurement
        and satellite observation in linear scale
    '''
    return sqrt(np.square(np.subtract(x,y)).mean())
```

```python
def NRMSD(x, y):
    ''' 
    Function to compute the normalized-root-mean-squared-deviation (NRMSD)
    
    Parameters:
    x (list): in-situ measurements [Linear scale]
    y (list): satellite observation [Linear scale]
    
    Returns:
    NRMSD (float): NRMSD value between in-situ measurement
        and satellite observation in linear scale
    '''
    return sqrt(np.square(np.subtract(x,y)).mean()) / np.nanmean(x)
```

```python
def MedPD(x, y):
    ''' 
    Function to compute the median-percentage-deviation (MedPD)
    
    Parameters:
    x (list): in-situ measurements [Linear scale]
    y (list): satellite observation [Linear scale]
    
    Returns:
    MedPD (float): MedPD value between in-situ measurement
        and satellite observation in linear scale
    '''
    return np.median(abs(y - x) / x * 100)
```

```python
def update_stats(matchup_stats, p, data, x, y, f, m, b, r2):
    '''
    Function to save the analysis statistics to the xr.Dataset
        
    Parameters:
    matchup_stats (xr.Dataset): empty dataset to store the regression analysis stasts.
    p (str): PFT or TChla name to store the data
    data (str): Name of the satellite dataset (e.g. DINEOF[transferred]) to store the statistics
    x (list): in-situ measurements [Linear scale]
    y (list): satellite observation [Linear scale]
    f (list): list of valid points that are finite in in-situ and satellite data
    m (float): slope of the regression
    b (float): intercept of the regression
    r2 (float): coefficient of determination of the regression

    Returns:
    matchup_stats (xr.Dataset): updated dataset with the matchup stasts saved
    '''
    
    matchup_stats[p].loc[data,'N. matchups'] = np.sum(f)
    matchup_stats[p].loc[data,'Slope'] = np.round(m,2)
    matchup_stats[p].loc[data,'Intercept'] = np.round(b,2)
    matchup_stats[p].loc[data,'R2'] = np.round(r2,2)
    matchup_stats[p].loc[data,'MedPD'] = np.round(MedPD(x[f],y[f]))
    matchup_stats[p].loc[data,'RMSD'] = np.round(RMSD(x[f],y[f]),2)
    matchup_stats[p].loc[data,'NRMSD'] = np.round(NRMSD(x[f],y[f]),2)
    
    return matchup_stats
```
